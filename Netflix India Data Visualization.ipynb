```python
ind.shape
```

```python
dr=ind['director'].value_counts()
dr
```

```python
dr1 = ind[ind['director'].isin(dr.index[dr > 5])]
dr1
```

```python
dr1.shape
```

```python
dr1 = dr1[dr1['director'] != 'Not Given']
dr1.shape
```

```python
plt.figure(figsize = (12,4))
sns.countplot(x = 'director', data=dr1)
plt.title('Count of Movies by Directors')
plt.xlabel('Name of Directors')
plt.ylabel('Number of Movies')
plt.xticks(rotation = 90)
plt.grid(ls=':', c = 'k')
plt.show()
```

Conclusion: 
1. From the above graph, we calculated the number of movies directed by the specified directors.
2. Highest number of movies was directed by 'Rajiv Chilaka', followed by 'Suhas Kadav'.

```python
plt.figure(figsize=(8,4))
sns.boxplot(x = 'release_year', y = 'director', data = dr1)
plt.title('Prime time of Directors')
plt.xlabel('Year of Release')
plt.ylabel('Name of Directors')
plt.grid(ls=':', c='k')
plt.show()
```

Conclusion:
1. Sooraj R Barjatya has a directing career started before 1990 and existed till 2015, serving for more than 25 years.
2. Most of the movies directed by David Dhawan was between year 2000 to 2013.
3. Rajkumar Santoshi, Ashutosh Gowariker and Ram Gopal Verma are having a good count of movies directed.

```python
dr1['duration']= dr1['duration'].str.replace(' min','')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = 'duration', y = 'director', data = dr1)
plt.title ('Duration of movies by director' )
plt.grid(ls=':',c='k')
plt.xlabel('Duration in minutes')
plt.ylabel('Name of Directors')
plt.show()
```

```python
plt.figure(figsize=(12,6))
dr1['rating'].value_counts().plot(kind='pie', autopct='%0.2f%%')
plt.title('Rating Chart of Movies')
plt.ylabel('Rating')
plt.show()
```

Conclusion:
1. TV-14 (Parents Strongly Cautioned) rated movies are having the highest count with 35.23%.
2. From the above pie chart, we can conclude that TV-Y7 (Directed to Older Children) has the second highest count.
3. TV-PG (Parental Guidance Suggested): 11.36%
4. TV-MA (Mature Audience): 7.95%
